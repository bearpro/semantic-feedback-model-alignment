# Project

This notebook converts existing `artifacts/infer/**/final.cs` files into tabular alignment artifacts: `model_index.csv`, `elements.csv`, `column_maps`, and projection CSVs for `property`, `type`, and `relation` layers.

```python
import pandas as pd
from IPython.display import JSON, display

from src.lib.align_pipeline import (
    load_column_map,
    load_projection_csv,
    prepare_project_artifacts,
)
from src.lib.infer_pipeline import find_repo_root

pd.set_option("display.max_rows", None)
pd.set_option("display.max_colwidth", None)
repo_root = find_repo_root()
```

```python
def project_progress(message: str) -> None:
    print(f"[project] {message}", flush=True)


project_summary, model_index, elements = prepare_project_artifacts(
    repo_root,
    progress_callback=project_progress,
)
display(JSON(project_summary, expanded=True))
```

```python
sample_model_id = str(model_index.iloc[0]["model_id"]) if not model_index.empty else None

if sample_model_id is None:
    display({"sample_model_id": None})
else:
    sample_property_projection = load_projection_csv(
        repo_root,
        "property",
        "path_plus_metadata_values",
        sample_model_id,
    )
    sample_property_column_map = load_column_map(
        repo_root,
        "property",
        sample_model_id,
    )
    sample_type_projection = load_projection_csv(
        repo_root,
        "type",
        "members_as_values",
        sample_model_id,
    )
    sample_relation_projection = load_projection_csv(
        repo_root,
        "relation",
        "path_plus_metadata_values",
        sample_model_id,
    )

    display(
        pd.DataFrame(
            [
                {
                    "sample_model_id": sample_model_id,
                    "property_columns": len(sample_property_projection.columns),
                    "property_column_map_entries": len(sample_property_column_map),
                    "type_columns": len(sample_type_projection.columns),
                    "relation_columns": len(sample_relation_projection.columns),
                }
            ]
        ).head()
    )
```
