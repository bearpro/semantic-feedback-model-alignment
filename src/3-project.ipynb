# Project

This notebook converts existing `artifacts/infer/**/final.cs` files into tabular alignment artifacts: `model_index.csv`, `elements.csv`, `column_maps`, and projection CSVs for `property`, `type`, and `relation` layers.

```python
import pandas as pd
from IPython.display import JSON, display

from src.lib.align_pipeline import (
    load_column_map,
    load_projection_csv,
    prepare_project_artifacts,
)
from src.lib.infer_pipeline import find_repo_root

pd.set_option("display.max_rows", None)
pd.set_option("display.max_colwidth", None)
repo_root = find_repo_root()
```

```python
project_summary, model_index, elements = prepare_project_artifacts(repo_root)
display(JSON(project_summary, expanded=True))
display(model_index.head(10))
display(elements.head(20))
```

```python
sample_model_id = str(model_index.iloc[0]["model_id"])

sample_property_projection = load_projection_csv(
    repo_root,
    "property",
    "path_plus_metadata_values",
    sample_model_id,
)
sample_property_column_map = load_column_map(
    repo_root,
    "property",
    sample_model_id,
)
sample_type_projection = load_projection_csv(
    repo_root,
    "type",
    "members_as_values",
    sample_model_id,
)
sample_relation_projection = load_projection_csv(
    repo_root,
    "relation",
    "path_plus_metadata_values",
    sample_model_id,
)

display({"sample_model_id": sample_model_id})
display(sample_property_projection)
display(JSON(sample_property_column_map, expanded=False))
display(sample_type_projection)
display(sample_relation_projection)
```
